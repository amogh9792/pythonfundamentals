Extract unique values of table column

```python
query = """
SELECT DISTINCT CountryCode 
FROM CountryLanguage
"""

data = pd.read_sql_query(query, myconn)
data
```

```python
data['CountryCode'].value_counts()
```

```python
query = """
SELECT *
FROM CountryLanguage
WHERE Isofficial = 'T'
"""

data = pd.read_sql_query(query, myconn)
data
```

```python
data['IsOfficial'].unique()
```

Extract records based on mathematical condition

```python
query = """
SELECT *
FROM CountryLanguage
WHERE Percentage > 75
"""

data = pd.read_sql_query(query, myconn)
data
```

Extract data based on two conditions

```python
query = """
SELECT *
FROM CountryLanguage
WHERE Percentage > 75
AND IsOfficial = 'T'
"""

data = pd.read_sql_query(query, myconn)
data
```

```python
query = """
SELECT *
FROM CountryLanguage
WHERE Percentage > 75
OR IsOfficial = 'T'
"""

data = pd.read_sql_query(query, myconn)
data
```

Read Data from table 'city' where country in 'IND' and city is 'Delhi'

```python
query = """
SELECT *
FROM city
WHERE CountryCode = 'IND'
AND Name = 'Delhi'
"""

data = pd.read_sql_query(query, myconn)
data
```

```python
my_city_list = ['Delhi', 'Mumbai', 'Chennai']

query = """
SELECT *
FROM city
WHERE CountryCode = 'IND'
AND Name IN {tuple(my_city_list)}
"""

data = pd.read_sql_query(query, myconn)
data
```

Get result in sorted form

```python
query = """
SELECT *
FROM city
ORDER BY Population ASC
"""

data = pd.read_sql_query(query, myconn)
data
```

Read data from table 'city' with name is ascending order and population in descending order.

```python
query = """
SELECT *
FROM city
ORDER BY Name ASC, Population DESC
"""

data = pd.read_sql_query(query, myconn)
data
```
